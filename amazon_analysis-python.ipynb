```python
# Importing necessary libraries
import pandas as pd

# Loading the dataset
df = pd.read_csv("amazon.csv")

# Displaying the first 5 rows to inspect the raw data
df.head()
```

```python
# Checking the data types and non-null counts
df.info()
```

```python
# Getting statistical summary of the data
df.describe()
```

```python
# Checking for missing values in each column
df.isnull().sum()
```

```python
# Handling missing values in 'rating_count' by filling with 0
df['rating_count'] = df['rating_count'].fillna(0)

# Verifying there are no more nulls
df.isnull().sum()
```

```python
# Standardizing text columns: converting to lowercase and stripping extra spaces
df['category'] = df['category'].str.lower().str.strip()
df['product_name'] = df['product_name'].str.strip()

# Filling missing reviews with a placeholder text
df['review_content'] = df['review_content'].fillna('no review')

# Checking the cleaned data
df.head()
```

```python
# --- PRICE RANGE FEATURE ---
# Cleaning 'actual_price': removing currency symbol and commas
df['actual_price'] = (
    df['actual_price']
    .astype(str)
    .str.replace('₹', '', regex=False)
    .str.replace(',', '', regex=False)
)

# Converting to numeric type
df['actual_price'] = pd.to_numeric(df['actual_price'], errors='coerce')

# Creating 'price_range' using quantiles (5 bins)
df['price_range'] = pd.qcut(
    df['actual_price'],
    q=5,
    labels=['very cheap', 'cheap', 'mid-range', 'expensive', 'premium']
)
```

```python
# --- DISCOUNT LEVEL FEATURE ---
# Cleaning 'discount_percentage': removing '%' symbol
df['discount_percentage'] = (
    df['discount_percentage']
    .str.replace('%', '', regex=False)
    .astype(float)
)

# Creating 'discount_level' using fixed bins
bins = [0, 10, 30, 50, 100]
labels = ['Low', 'Medium', 'High', 'Very High']

df['discount_level'] = pd.cut(
    df['discount_percentage'],
    bins=bins,
    labels=labels,
    include_lowest=True
)
```

```python
# --- ENGAGEMENT SCORE CALCULATION ---
# Converting 'rating' to numeric
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Cleaning 'rating_count': removing commas and plus signs
df['rating_count'] = (
    df['rating_count']
    .astype(str)
    .str.replace(',', '', regex=False)
    .str.replace('+', '', regex=False)
)

# Converting 'rating_count' to numeric and ensuring no nulls
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')
df['rating_count'] = df['rating_count'].fillna(0)

# Calculating the Engagement Score (Rating * Rating Count)
df['engagement_score'] = df['rating'] * df['rating_count']

# Previewing the new metrics
df[['rating', 'rating_count', 'engagement_score']].head(5)
```

```python
# Creating 'engagement_level' based on the score quantiles
df['engagement_level'] = pd.qcut(
    df['engagement_score'],
    q=4,
    labels=['Low', 'Medium', 'High', 'Very High']
)
```

```python
# Analysis: Average Engagement by Price Range
df.groupby('price_range')['engagement_score'].mean().sort_values(ascending=False)
```

```python
# Analysis: Average Engagement by Discount Level
# This confirms our hypothesis about 'Medium' discounts
df.groupby('discount_level', observed=True)['engagement_score'] \
  .mean() \
  .sort_values(ascending=False)
```

```python
# Visualizing the Engagement by Discount Level
df.groupby('discount_level', observed=True)['engagement_score'] \
  .mean() \
  .plot(kind='bar', title='Engagement by Discount Level')
```

```python
# Exporting the cleaned dataframe to CSV (for backup)
df.to_csv("amazon_updated.csv", index=False)

# Exporting to Excel (optional, for reporting)
df.to_excel("amazon_updated.xlsx", index=False)

print("Data exported successfully.")
```

```python
# --- STEP 1: IMPORT LIBRARIES ---
# (Run 'pip install psycopg2-binary sqlalchemy' in your terminal if not installed)
from sqlalchemy import create_engine, text

# --- STEP 2: DATABASE CONNECTION ---
username = 'postgres' 
password = 'YOUR_PASSWORD_HERE'
host = 'localhost'
port = '5432'
database = 'customer_analysis'

# Create the engine
engine = create_engine(f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}")

print("Engine configuration set.")
```

```python
# --- STEP 3: CLEAN SLATE ---
# This drops the old table and views to ensure a fresh upload
with engine.connect() as conn:
    conn.execute(text("DROP VIEW IF EXISTS public.product_dashboard CASCADE"))
    conn.execute(text("DROP VIEW IF EXISTS public.customer_engagement_summary CASCADE"))
    conn.execute(text("DROP TABLE IF EXISTS public.customer CASCADE"))
    conn.commit() 
    print("✅ Old tables and views deleted.")
```

```python
# --- STEP 4: UPLOAD DATA ---
# Uploading the cleaned dataframe ('df') to PostgreSQL
df.to_sql("customer", engine, if_exists="replace", index=False)
print("✅ Data uploaded to 'customer' table successfully.")
```

```python
# --- STEP 5: RECREATE DASHBOARD VIEW ---
# This view is optimized for Power BI to read the data easily
with engine.connect() as conn:
    conn.execute(text("""
        CREATE OR REPLACE VIEW public.product_dashboard AS
        SELECT
            product_id,
            product_name,
            SPLIT_PART(category, '|', 1) AS main_category,
            discounted_price,
            discount_level,
            price_range,
            rating::NUMERIC AS rating,
            rating_count::NUMERIC AS rating_count,
            engagement_score
        FROM customer
        WHERE rating IS NOT NULL
          AND engagement_score IS NOT NULL;
    """))
    conn.commit()
    print("✅ Dashboard view created. Power BI is ready.")
```
